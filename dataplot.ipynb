```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_1=np.loadtxt("Ye_test1.dat")
x1=file_1[:,0]
y1=file_1[:,1]
z1=file_1[:,2]
```

```python
file_2=np.loadtxt("Ye_test2.dat")
x2=file_2[:,0]
y2=file_2[:,1]
z2=file_2[:,2]
```

```python
plt.figure(figsize=[10,5])
plt.plot(x1,y1,'-r',label="curve of second column vs.first column")
plt.plot(x1,z1,'-b',label="curve of third column vs.first column")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[10,5])
plt.plot(x2,y2,'-r',label="curve of second column vs.first column")
plt.plot(x2,z2,'-b',label="curve of third column vs.first column")
plt.legend()
plt.show()
```
